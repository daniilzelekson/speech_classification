```
import numpy as np

def cohen_d(group1, group2):
    diff = group1.mean() - group2.mean()
    n1, n2 = len(group1), len(group2)
    var1 = group1.var()
    var2 = group2.var()
    pooled_var = (n1 * var1 + n2 * var2) / (n1 + n2)
    d = diff / np.sqrt(pooled_var)
    return d

# создаем две группы
group_patient = df[df['group'] == 'patient']['quantity_words_in_sec']
group_control = df[df['group'] == 'control']['quantity_words_in_sec']

# расчет размера эффекта
d = cohen_d(group_patient, group_control)

print(f"Cohen's d: {d}")
```

```
from gensim.models import Word2Vec, FastText
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Токенизация текстов
tokenized_texts = df['text'].apply(lambda x: x.split())

# Обучение моделей Word2Vec и FastText
w2v_model = Word2Vec(tokenized_texts, min_count=1)
ft_model = FastText(tokenized_texts, min_count=1)

# Получение векторов слов
w2v_vectors = w2v_model.wv
ft_vectors = ft_model.wv

# Получение среднего вектора для каждого текста
w2v_matrix = np.array([np.mean([w2v_vectors[w] for w in text if w in w2v_vectors], axis=0) for text in tokenized_texts])
ft_matrix = np.array([np.mean([ft_vectors[w] for w in text if w in ft_vectors], axis=0) for text in tokenized_texts])

# Применение PCA для сокращения размерности до 2D и 3D для визуализации
pca_2d = PCA(n_components=2)
w2v_2d = pca_2d.fit_transform(w2v_matrix)
ft_2d = pca_2d.fit_transform(ft_matrix)

pca_3d = PCA(n_components=3)
w2v_3d = pca_3d.fit_transform(w2v_matrix)
ft_3d = pca_3d.fit_transform(ft_matrix)

# Визуализация 2D и 3D
diagnosis_colors = {'healthy': 'blue', 'depressive': 'red', 'bipolar': 'green', 'schizophrenia': 'purple'}
for name, matrix_2d, matrix_3d in [('Word2Vec', w2v_2d, w2v_3d), ('FastText', ft_2d, ft_3d)]:
    # 2D
    plt.figure(figsize=(10, 10))
    for diagnosis in diagnosis_colors.keys():
        indices = df['diagnosis'] == diagnosis
        plt.scatter(matrix_2d[indices, 0], matrix_2d[indices, 1], color=diagnosis_colors[diagnosis], label=diagnosis)
    plt.legend()
    plt.title(f'{name} 2D')
    plt.show()

    # 3D
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    for diagnosis in diagnosis_colors.keys():
        indices = df['diagnosis'] == diagnosis
        ax.scatter(matrix_3d[indices, 0], matrix_3d[indices, 1], matrix_3d[indices, 2], color=diagnosis_colors[diagnosis], label=diagnosis)
    ax.legend()
    plt.title(f'{name} 3D')
    plt.show()
```

```
import spacy
import matplotlib.pyplot as plt
import seaborn as sns

# Load the spacy model
nlp = spacy.load('ru_core_news_sm')

# Function to count parts of speech
def count_pos(text):
    doc = nlp(text)
    pos_counts = {}
    for token in doc:
        if token.pos_ in pos_counts:
            pos_counts[token.pos_] += 1
        else:
            pos_counts[token.pos_] = 1
    return pos_counts

# Apply the function to the text column
df['pos_counts'] = df['text'].apply(count_pos)

# Create a dataframe for visualization
vis_df = df['pos_counts'].apply(pd.Series).fillna(0)
vis_df['diagnosis'] = df['diagnosis']

# Melt the dataframe for visualization
melted_df = pd.melt(vis_df, id_vars=['diagnosis'], var_name='Part of Speech', value_name='Count')

# Visualize the counts of parts of speech
plt.figure(figsize=(10, 6))
sns.barplot(x='Part of Speech', y='Count', hue='diagnosis', data=melted_df, ci=None)
plt.title('Counts of Parts of Speech by Diagnosis')
plt.show()
```
